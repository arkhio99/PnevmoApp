```python
from keras import layers
from keras import models
import os, sys
from PIL import Image, ImageOps
```

```python
import pandas as pd
import numpy as np
```

## Подготовка данных

Прочитаем данные

```python
train_folder = "../data1/train"
test_folder = "../data1/test"
valid_folder = "../data1/val"
ind_by_label = {
    'normal': 0,
    'opacity': 1
}

new_size = (256, 256)

def image_to_matrix(im, resize_size):
    im = ImageOps.grayscale(im.resize(resize_size))
    (w, h) = im.size
    im_pixs = np.array(list(im.getdata())).reshape((w, h)).astype('float32')/255
    return im_pixs

def get_dataset(data_folder_name, folder_name):
    data = pd.DataFrame()
    image_names = os.listdir(data_folder_name + '\\' + folder_name)
    c = 0.0
    for image_name in image_names:
        im = Image.open(data_folder_name + '\\' + folder_name + '\\' + image_name)
        im_pixs = image_to_matrix(im, new_size)
        im_data = pd.Series(
            {
                'Data': im_pixs,
                'Label': ind_by_label[folder_name],
            }
        )
        data = data.append(im_data, ignore_index=True)
        c += 1.0
        print(c)
        print(c / len(image_names))
    
    return data
```

```python
train_normal_data = get_dataset(train_folder, 'normal')
test_normal_data = get_dataset(test_folder, 'normal')
```

```python
val_normal_data = get_dataset(valid_folder, 'normal')
```

```python
train_opacity_data = get_dataset(train_folder, 'opacity')
test_opacity_data = get_dataset(test_folder, 'opacity')
val_opacity_data = get_dataset(valid_folder, 'opacity')
```

```python
from sklearn.utils import resample

min_train_len = np.array([len(train_normal_data), len(train_opacity_data)]).min()

train_resample_normal_data = resample(train_normal_data, replace=True, n_samples=min_train_len, random_state=42)
train_resample_opacity_data = resample(train_opacity_data, replace=True, n_samples=min_train_len, random_state=42)
```

```python
from sklearn.utils import resample

min_test_len = np.array([len(test_normal_data), len(test_opacity_data)]).min()

test_resample_normal_data = resample(test_normal_data, replace=True, n_samples=min_test_len, random_state=42)
test_resample_opacity_data = resample(test_opacity_data, replace=True, n_samples=min_test_len, random_state=42)
```

```python
from sklearn.utils import resample

min_val_len = np.array([len(val_normal_data), len(val_opacity_data)]).min()

val_resample_normal_data = resample(val_normal_data, replace=True, n_samples=min_val_len, random_state=42)
val_resample_opacity_data = resample(val_opacity_data, replace=True, n_samples=min_val_len, random_state=42)
```

```python
def shuffle(X, y):
    df = pd.DataFrame({
        "Data" : X,
        "Label" : y
    })
    df = df.sample(frac=1)
    X = df["Data"]
    y = df["Label"]
    return (X, y)
```

```python
X_train = train_resample_normal_data['Data'] \
    .append(train_resample_opacity_data['Data'])
y_train = train_resample_normal_data['Label'] \
    .append(train_resample_opacity_data['Label'])
(X_train, y_train) = shuffle(X_train, y_train)

X_test = test_resample_normal_data['Data'] \
    .append(test_resample_opacity_data['Data'])
y_test = test_resample_normal_data['Label'] \
    .append(test_resample_opacity_data['Label'])
(X_test, y_test) = shuffle(X_test, y_test)

X_val = val_resample_normal_data['Data'] \
    .append(val_resample_opacity_data['Data'])
y_val = val_resample_normal_data['Label'] \
    .append(val_resample_opacity_data['Label'])
(X_val, y_val) = shuffle(X_val, y_val)
```

```python
from tensorflow.keras.utils import to_categorical
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
y_val_cat = to_categorical(y_val)
```

```python
counts= [0, 0]
for i in range(0, y_train.shape[0]):
    counts[int(y_train_cat[i][0])]+=1
counts
```

## Создание моделей

### Свёрточная нейросеть

```python
cnn = models.Sequential()
cnn.add(layers.Conv2D(10, (3, 3), activation='relu', input_shape=(new_size[0], new_size[1], 1)))
cnn.add(layers.MaxPool2D((3, 3)))
cnn.add(layers.Conv2D(5, (5, 5), activation='relu'))
cnn.add(layers.MaxPool2D((7, 7)))
#cnn.add(layers.Conv3D(1, (10, 10, 1), activation='relu'))
#cnn.add(layers.MaxPool3D((3, 3, 1)))
cnn.add(layers.Flatten())
cnn.add(layers.Dense(12, activation='relu'))
cnn.add(layers.Dense(6, activation='relu'))
cnn.add(layers.Dense(2, activation='softmax'))
```

```python
cnn.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'])
```

```python
from tensorflow import convert_to_tensor
```

```python
x_train_cnn = [convert_to_tensor(x) for x in X_train]
x_val_cnn = [convert_to_tensor(x) for x in X_val]
```

```python
cnn.fit(np.array(x_train_cnn), np.array(y_train_cat), epochs=6, batch_size=15, validation_data=(np.array(x_val_cnn), np.array(y_val_cat)))
```

Проверим на тестовых данных

```python
x_test_cnn = [convert_to_tensor(x) for x in X_test]
```

```python
test_cnn_loss, test_cnn_acc = cnn.evaluate(np.array(x_test_cnn), np.array(y_test_cat))
```
