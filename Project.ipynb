```python
from keras import layers
from keras import models
import os, sys
from PIL import Image, ImageOps
```

```python
import pandas as pd
import numpy as np
```

## Подготовка данных

Прочитаем данные

```python
data_folder = "data"
ind_by_label = {
    'normal': 0,
    'opacity': 1
}

new_size = (256, 256)

def image_to_matrix(im, resize_size):
    im = ImageOps.grayscale(im.resize(resize_size))
    (w, h) = im.size
    im_pixs = np.array(list(im.getdata())).reshape((w, h)).astype('float32')/255
    return im_pixs

def get_dataset(data_folder_name, folder_name):
    data = pd.DataFrame()
    image_names = os.listdir(data_folder_name + '\\' + folder_name)
    c = 0.0
    for image_name in image_names:
        im = Image.open(data_folder_name + '\\' + folder_name + '\\' + image_name)
        im_pixs = image_to_matrix(im, new_size)
        im_data = pd.Series(
            {
                'Data': im_pixs,
                'Label': ind_by_label[folder_name],
            }
        )
        data = data.append(im_data, ignore_index=True)
        c += 1.0
        print(c)
        print(c / len(image_names))
    
    return data
```

```python
normal_data = get_dataset(data_folder, 'normal')
```

```python
opacity_data = get_dataset(data_folder, 'opacity')
```

```python
from sklearn.utils import resample

min_len = np.array([len(normal_data), len(opacity_data)]).min()

normal_data = resample(normal_data, replace=True, n_samples=min_len, random_state=42)
opacity_data = resample(opacity_data, replace=True, n_samples=min_len, random_state=42)
```

```python
X = normal_data['Data'] \
    .append(opacity_data['Data'])
y = normal_data['Label'] \
    .append(opacity_data['Label'])
```

```python
from tensorflow.keras.utils import to_categorical
y1 = to_categorical(y)
```

Разделим выборку на тренировочную и тестовую

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y1, shuffle=True, test_size=0.25)
```

```python
val_count = 50
x_val = x_train[-val_count:]
y_val = y_train[-val_count:]
x_train = x_train[:-val_count]
y_train = y_train[:-val_count]
```

## Создание моделей

### Свёрточная нейросеть

```python
cnn = models.Sequential()
cnn.add(layers.Conv2D(5, (15, 15), activation='relu', input_shape=(new_size[0], new_size[1], 1)))
cnn.add(layers.MaxPool2D((3, 3)))
#cnn.add(layers.Conv3D(1, (10, 10, 1), activation='relu'))
#cnn.add(layers.MaxPool3D((3, 3, 1)))
cnn.add(layers.Flatten())
#cnn.add(layers.Dense(9, activation='relu'))
cnn.add(layers.Dense(2, activation='softmax'))
```

```python
cnn.compile(optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['accuracy'])
```

```python
from tensorflow import convert_to_tensor
```

```python
x_train_cnn = [convert_to_tensor(x) for x in x_train]
x_val_cnn = [convert_to_tensor(x) for x in x_val]
```

```python
cnn.fit(np.array(x_train_cnn), np.array(y_train), epochs=5, batch_size=25, validation_data=(np.array(x_val_cnn), np.array(y_val)))
```

Проверим на тестовых данных

```python
x_test_cnn = [convert_to_tensor(x) for x in x_test]
```

```python
test_cnn_loss, test_cnn_acc = cnn.evaluate(np.array(x_test_cnn), np.array(y_test))
```
